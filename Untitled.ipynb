```python
%matplotlib inline
import numpy as np
import matplotlib as mlp
import matplotlib.pyplot as plt
np.random.seed(sum(map(ord, "aesthetics")))
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 10000)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)

sinplot()
```
